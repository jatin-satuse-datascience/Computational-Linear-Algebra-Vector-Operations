```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([2, 3])
b = np.array([1, 4])


# Scalar multiplication of a
scalar = 2
result_scalar = scalar * a
print(f"Scalar multiplication of a ({scalar} * a):", result_scalar)

# Plot scalar multiplication of a
plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='g',label='a')
plt.quiver(0,0,result_scalar[0],result_scalar[1],angles='xy',scale_units='xy',scale=1,color='m',label=f'{scalar}*a')

plt.xlim(0,6)
plt.ylim(0,10)
plt.legend()
plt.show()



# Scalar Multiplication of b
scalar = 3
result_scalar_b= scalar * b
print (f"Scalar Multiplication of b = ({scalar} * b ):", result_scalar_b)


# Plot scalar multiplication of b
plt.quiver(0,0, b[0], b[1],angles = 'xy',scale_units='xy',scale=1,color= 'g', label='b' )
plt.quiver(0,0,result_scalar_b[0],result_scalar_b[1],angles='xy',scale_units='xy',scale=1,color='r',label=f'{scalar}*b')

plt.xlim(0,6)
plt.ylim(0,15)
plt.legend()
plt.show()
```
