```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([2, 3])
b = np.array([1, 4])
result = a + b
print ("Vector a:",a)
print ("Vector b:",b)
print (result)
```

```python
plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='r',label='a')
plt.quiver(0,0,b[0],b[1],angles='xy',scale_units='xy',scale=1,color='g',label='b')
plt.quiver(0,0,result[0],result[1],angles='xy',scale_units='xy',scale=1,color='b',label='a+b')

plt.xlim(0,5)
plt.ylim(0,9)
plt.legend()
plt.show()
```
