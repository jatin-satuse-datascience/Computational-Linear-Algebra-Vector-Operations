```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define two 3D vectors
A = np.array([2, 3, 4])
B = np.array([5, 6, 7])

# Compute the cross product
C = np.cross(A, B)

#Printing Numerical Output
print("Vector A:", A)
print("Vector B:", B)
print("Cross Product (A × B):", C)


# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Origin
origin = np.array([0, 0, 0])

# Plot vectors
ax.quiver(*origin, *A, color='r', label='Vector A')
ax.quiver(*origin, *B, color='b', label='Vector B')
ax.quiver(*origin, *C, color='g', label='Cross Product (A × B)')

# Labels and legend
ax.set_xlim([0, 8])
ax.set_ylim([0, 8])
ax.set_zlim([0, 8])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
```
